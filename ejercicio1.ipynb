```python
#create an instance of the class using our radar
display = pyart.graph.RadarMapDisplay(RadParStrat)

#me quedo con una sola elevacion
radar2 = RadParStrat.extract_sweeps([0])

#create a Matplotlib figure
f = plt.figure(figsize = [17,4])
minlon=radar2.longitude['data']-1.
maxlon=radar2.longitude['data']+1.
minlat=radar2.latitude['data']-1.
maxlat=radar2.latitude['data']+1.


#now we are going to do a three panel plot, resolution is a basemap parameter and determines the resolution of 
#the coastline.. here we set to intermediate or 'i' ('h' for high 'l' for low)

#plt.subplot(1, 3, 1) 
#display.plot_ppi('ZDR',vmin = -7, vmax = 7)
plt.subplot(1, 3, 2) 
display.plot_ppi_map('dBZ', max_lat = maxlat, min_lat =minlat, min_lon = minlon, max_lon = maxlon,
                     vmin = -8, vmax = 64, lat_lines = np.arange(minlat[0],maxlat[0],.2), lon_lines = np.arange(minlon[0], maxlon[0], .9),
                     resolution = 'i')
#plt.subplot(1, 3, 3) 
#display.plot_ppi('V', max_lat = maxlat, min_lat =minlat, min_lon = minlon, max_lon = maxlon,
#                     vmin = -15, vmax = 15, lat_lines = np.arange(-30,-40,.2), lon_lines = np.arange(-100, -90, .5),
#                     resolution = 'i')
```
